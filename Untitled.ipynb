```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snc
from sklearn.datasets import load_boston
#from sklearn.datasets import boston_price
```

```python
boston=load_boston()
boston
```

```python
boston.keys()
```

```python
df_boston=pd.DataFrame(boston.data, columns=boston.feature_names)
df_boston.head()
```

```python
df_boston.describe()
```

```python
snc.heatmap(df_boston.corr(),annot=True)
```

```python
df_boston['MEDV']=boston.target
df_boston.head()
```

```python
df_boston.corr()
```

```python
plt.scatter(df_boston['MEDV'],df_boston["B"])
plt.xlabel("MEDV")
plt.ylabel("B")
plt.show()
```

```python
plt.scatter(df_boston['MEDV'],df_boston["CRIM"])
plt.xlabel("MEDV")
plt.ylabel("CRIM")
plt.show()

```

```python
plt.scatter(df_boston['MEDV'],df_boston["NOX"])
plt.xlabel("MEDV")
plt.ylabel("NOX")
plt.show()
```

```python
plt.scatter(df_boston['MEDV'],df_boston['LSTAT'])
plt.xlabel('MEDV')
plt.ylabel('LSTAT')
plt.show()
```

WE can see that the relation between (LSTAt) & (MEDV) is inverse

```python
plt.scatter(df_boston['MEDV'],df_boston['RM'])
plt.xlabel('MEDV')
plt.ylabel('RM')
plt.show()
```

WE can see that the relation between (RM) & (MEDV) is direct

As we saw from the correlation and from scatter plot, (RM) & (LSTAT) plays a big role in (MEDV)
